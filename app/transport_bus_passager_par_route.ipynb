```python
import pandas as pd
import plotly.graph_objects as go
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
from fbprophet import Prophet

# ==============================
# üîπ √âtape 1 : Chargement des donn√©es
# ==============================
data = pd.read_csv('data/transport_bus_passager_par_route.csv', sep=';')

# Transformation du DataFrame pour Prophet
df_melted = pd.melt(data, id_vars='Route', var_name='Ann√©e', value_name='Valeur')

# Extraction de l'ann√©e (ex: "1999/00" ‚Üí "1999")
df_melted['Ann√©e'] = df_melted['Ann√©e'].str[:4]
df_melted['Ann√©e'] = pd.to_datetime(df_melted['Ann√©e'], format='%Y')

# Tri des donn√©es
df_melted = df_melted.sort_values(by=['Route', 'Ann√©e']).reset_index(drop=True)

# ==============================
# üîπ √âtape 2 : Fonction de pr√©diction
# ==============================
def predict_route(selected_route):
    df_filtered = df_melted[df_melted["Route"] == selected_route]

    # Cr√©ation du DataFrame pour Prophet
    df_prophet = df_filtered[['Ann√©e', 'Valeur']].rename(columns={'Ann√©e': 'ds', 'Valeur': 'y'})

    # Initialisation et entra√Ænement de Prophet
    model = Prophet(interval_width=0.95)
    model.fit(df_prophet)

    # Cr√©ation du DataFrame pour les pr√©visions (jusqu'en 2024)
    future = model.make_future_dataframe(periods=11, freq='Y')  # 11 ans apr√®s 2013
    forecast = model.predict(future)

    return df_prophet, forecast

# ==============================
# üîπ √âtape 3 : Application Dash
# ==============================
app = dash.Dash(__name__)

# Layout de l'application
app.layout = html.Div([
    html.Label("S√©lectionnez une Route :"),
    dcc.Dropdown(
        id='route-dropdown',
        options=[{'label': f"Route {route}", 'value': route} for route in df_melted["Route"].unique()],
        value=df_melted["Route"].unique()[0],  # Route s√©lectionn√©e par d√©faut
        style={'width': '50%'}
    ),
    dcc.Graph(id='evolution-graph')
])

# Callback pour mettre √† jour le graphique
@app.callback(
    Output('evolution-graph', 'figure'),
    [Input('route-dropdown', 'value')]
)
def update_graph(selected_route):
    df_prophet, forecast = predict_route(selected_route)

    fig = go.Figure()

    # üîπ Ajout des donn√©es historiques
    fig.add_trace(go.Scatter(
        x=df_prophet["ds"],
        y=df_prophet["y"],
        mode='lines+markers',
        name=f"Observ√© - Route {selected_route}",
        hovertemplate='%{y} v√©hicules<br>Ann√©e: %{x|%Y}<extra></extra>',
    ))

    # üîπ Ajout des pr√©visions
    fig.add_trace(go.Scatter(
        x=forecast["ds"],
        y=forecast["yhat"],
        mode='lines',
        name=f"Pr√©diction - Route {selected_route}",
        line=dict(dash='dash', color='red'),
        hovertemplate='%{y:.0f} v√©hicules<br>Ann√©e: %{x|%Y}<extra></extra>',
    ))

    # üîπ Ajout de l'incertitude
    fig.add_trace(go.Scatter(
        x=forecast["ds"].tolist() + forecast["ds"][::-1].tolist(),
        y=forecast["yhat_upper"].tolist() + forecast["yhat_lower"][::-1].tolist(),
        fill='toself',
        fillcolor='rgba(255, 0, 0, 0.2)',
        line=dict(color='rgba(255,255,255,0)'),
        hoverinfo="skip",
        showlegend=True,
        name="Incertitude",
    ))

    # üîπ Mise en forme du graphique
    fig.update_layout(
        title=f'√âvolution et Pr√©diction du trafic pour la Route {selected_route}',
        xaxis_title='Ann√©e',
        yaxis_title='Nombre de v√©hicules',
        hovermode='x unified',
        template='plotly_dark',
    )

    return fig

# ==============================
# üîπ √âtape 4 : Lancement de l'application
# ==============================
if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
# Importation des biblioth√®ques n√©cessaires
import pandas as pd
import plotly.graph_objects as go
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
from prophet import Prophet

# Lecture du fichier CSV
data = pd.read_csv('data/transport_bus_passager_par_route.csv', sep=';')

# Transformation du DataFrame pour Prophet
df_melted = pd.melt(data, 
                    id_vars='Route', 
                    var_name='Ann√©e', 
                    value_name='Valeur')

# Conversion du format de l'ann√©e
df_melted['Ann√©e'] = df_melted['Ann√©e'].str[:4]  # Garde juste le d√©but de l'ann√©e (ex: 1999)
df_melted['Ann√©e'] = pd.to_datetime(df_melted['Ann√©e'], format='%Y')
df_melted = df_melted.sort_values(by=['Route', 'Ann√©e']).reset_index(drop=True)

# Fonction de pr√©paration des donn√©es pour Prophet
def prepare_data_for_prophet(selected_route):
    df_filtered = df_melted[df_melted["Route"] == selected_route]
    df_prophet = df_filtered[["Ann√©e", "Valeur"]].rename(columns={"Ann√©e": "ds", "Valeur": "y"})
    return df_prophet

# Fonction de pr√©diction avec Prophet
def predict_route(selected_route):
    # Pr√©paration des donn√©es
    df_prophet = prepare_data_for_prophet(selected_route)

    # Initialisation de Prophet
    model = Prophet(interval_width=0.95)
    model.fit(df_prophet)

    # G√©n√©ration des dates futures jusqu'√† 2024
    future_dates = model.make_future_dataframe(periods=12, freq='Y')
    forecast = model.predict(future_dates)

    return df_prophet, forecast

# Initialisation de l'application Dash
app = dash.Dash(__name__)

# Layout de l'application Dash
app.layout = html.Div([
    html.Label("S√©lectionnez une Route :"),
    dcc.Dropdown(
        id='route-dropdown',
        options=[{'label': f"Route {route}", 'value': route} for route in df_melted["Route"].unique()],
        value=df_melted["Route"].unique()[0],  # Route s√©lectionn√©e par d√©faut
        style={'width': '50%'}
    ),

    dcc.Graph(id='evolution-graph')
])

# Callback pour mettre √† jour le graphique en fonction de la route s√©lectionn√©e
@app.callback(
    Output('evolution-graph', 'figure'),
    [Input('route-dropdown', 'value')]
)
def update_graph(selected_route):
    df_prophet, forecast = predict_route(selected_route)

    fig = go.Figure()

    # Tracer les donn√©es observ√©es
    fig.add_trace(go.Scatter(
        x=df_prophet["ds"],
        y=df_prophet["y"],
        mode='lines+markers',
        name=f"Observ√© - Route {selected_route}",
        hovertemplate='%{y} v√©hicules<br>Ann√©e: %{x|%Y}<extra></extra>',
    ))

    # Tracer les pr√©dictions
    fig.add_trace(go.Scatter(
        x=forecast["ds"],
        y=forecast["yhat"],
        mode='lines',
        name=f"Pr√©diction - Route {selected_route}",
        line=dict(dash='dash', color='red'),
        hovertemplate='%{y:.0f} v√©hicules<br>Ann√©e: %{x|%Y}<extra></extra>',
    ))

    # Ajouter l'incertitude (intervalle de confiance)
    fig.add_trace(go.Scatter(
        x=forecast["ds"].tolist() + forecast["ds"][::-1].tolist(),
        y=forecast["yhat_upper"].tolist() + forecast["yhat_lower"][::-1].tolist(),
        fill='toself',
        fillcolor='rgba(255, 0, 0, 0.2)',
        line=dict(color='rgba(255,255,255,0)'),
        hoverinfo="skip",
        showlegend=True,
        name="Incertitude",
    ))

    # Mise en page du graphique
    fig.update_layout(
        title=f'√âvolution et Pr√©diction du trafic pour la Route {selected_route}',
        yaxis_title='Nombre de v√©hicules',
        hovermode='x unified',
        template='plotly_dark',
    )

    return fig

# Lancer l'application
if __name__ == '__main__':
    app.run_server(debug=True)
```
