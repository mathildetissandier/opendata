```python
import pandas as pd
import plotly.graph_objects as go
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
from fbprophet import Prophet

# ==============================
# ðŸ”¹ Ã‰tape 1 : Chargement des donnÃ©es
# ==============================
data = pd.read_csv('data/transport_bus_passager_par_route.csv', sep=';')

# Transformation du DataFrame pour Prophet
df_melted = pd.melt(data, id_vars='Route', var_name='AnnÃ©e', value_name='Valeur')

# Extraction de l'annÃ©e (ex: "1999/00" â†’ "1999")
df_melted['AnnÃ©e'] = df_melted['AnnÃ©e'].str[:4]
df_melted['AnnÃ©e'] = pd.to_datetime(df_melted['AnnÃ©e'], format='%Y')

# Tri des donnÃ©es
df_melted = df_melted.sort_values(by=['Route', 'AnnÃ©e']).reset_index(drop=True)

# ==============================
# ðŸ”¹ Ã‰tape 2 : Fonction de prÃ©diction
# ==============================
def predict_route(selected_route):
    df_filtered = df_melted[df_melted["Route"] == selected_route]

    # CrÃ©ation du DataFrame pour Prophet
    df_prophet = df_filtered[['AnnÃ©e', 'Valeur']].rename(columns={'AnnÃ©e': 'ds', 'Valeur': 'y'})

    # Initialisation et entraÃ®nement de Prophet
    model = Prophet(interval_width=0.95)
    model.fit(df_prophet)

    # CrÃ©ation du DataFrame pour les prÃ©visions (jusqu'en 2024)
    future = model.make_future_dataframe(periods=11, freq='Y')  # 11 ans aprÃ¨s 2013
    forecast = model.predict(future)

    return df_prophet, forecast

# ==============================
# ðŸ”¹ Ã‰tape 3 : Application Dash
# ==============================
app = dash.Dash(__name__)

# Layout de l'application
app.layout = html.Div([
    html.Label("SÃ©lectionnez une Route :"),
    dcc.Dropdown(
        id='route-dropdown',
        options=[{'label': f"Route {route}", 'value': route} for route in df_melted["Route"].unique()],
        value=df_melted["Route"].unique()[0],  # Route sÃ©lectionnÃ©e par dÃ©faut
        style={'width': '50%'}
    ),
    dcc.Graph(id='evolution-graph')
])

# Callback pour mettre Ã  jour le graphique
@app.callback(
    Output('evolution-graph', 'figure'),
    [Input('route-dropdown', 'value')]
)
def update_graph(selected_route):
    df_prophet, forecast = predict_route(selected_route)

    fig = go.Figure()

    # ðŸ”¹ Ajout des donnÃ©es historiques
    fig.add_trace(go.Scatter(
        x=df_prophet["ds"],
        y=df_prophet["y"],
        mode='lines+markers',
        name=f"ObservÃ© - Route {selected_route}",
        hovertemplate='%{y} vÃ©hicules<br>AnnÃ©e: %{x|%Y}<extra></extra>',
    ))

    # ðŸ”¹ Ajout des prÃ©visions
    fig.add_trace(go.Scatter(
        x=forecast["ds"],
        y=forecast["yhat"],
        mode='lines',
        name=f"PrÃ©diction - Route {selected_route}",
        line=dict(dash='dash', color='red'),
        hovertemplate='%{y:.0f} vÃ©hicules<br>AnnÃ©e: %{x|%Y}<extra></extra>',
    ))

    # ðŸ”¹ Ajout de l'incertitude
    fig.add_trace(go.Scatter(
        x=forecast["ds"].tolist() + forecast["ds"][::-1].tolist(),
        y=forecast["yhat_upper"].tolist() + forecast["yhat_lower"][::-1].tolist(),
        fill='toself',
        fillcolor='rgba(255, 0, 0, 0.2)',
        line=dict(color='rgba(255,255,255,0)'),
        hoverinfo="skip",
        showlegend=True,
        name="Incertitude",
    ))

    # ðŸ”¹ Mise en forme du graphique
    fig.update_layout(
        title=f'Ã‰volution et PrÃ©diction du trafic pour la Route {selected_route}',
        xaxis_title='AnnÃ©e',
        yaxis_title='Nombre de vÃ©hicules',
        hovermode='x unified',
        template='plotly_dark',
    )

    return fig

# ==============================
# ðŸ”¹ Ã‰tape 4 : Lancement de l'application
# ==============================
if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
# Importation des bibliothÃ¨ques nÃ©cessaires
import pandas as pd
import plotly.graph_objects as go
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
from prophet import Prophet

# Lecture du fichier CSV
data = pd.read_csv('data/transport_bus_passager_par_route.csv', sep=';')

# Transformation du DataFrame pour Prophet
df_melted = pd.melt(data, 
                    id_vars='Route', 
                    var_name='AnnÃ©e', 
                    value_name='Valeur')

# Conversion du format de l'annÃ©e
df_melted['AnnÃ©e'] = df_melted['AnnÃ©e'].str[:4]  # Garde juste le dÃ©but de l'annÃ©e (ex: 1999)
df_melted['AnnÃ©e'] = pd.to_datetime(df_melted['AnnÃ©e'], format='%Y')
df_melted = df_melted.sort_values(by=['Route', 'AnnÃ©e']).reset_index(drop=True)

# Fonction de prÃ©paration des donnÃ©es pour Prophet
def prepare_data_for_prophet(selected_route):
    df_filtered = df_melted[df_melted["Route"] == selected_route]
    df_prophet = df_filtered[["AnnÃ©e", "Valeur"]].rename(columns={"AnnÃ©e": "ds", "Valeur": "y"})
    return df_prophet

# Fonction de prÃ©diction avec Prophet
def predict_route(selected_route):
    # PrÃ©paration des donnÃ©es
    df_prophet = prepare_data_for_prophet(selected_route)

    # Initialisation de Prophet
    model = Prophet(interval_width=0.95)
    model.fit(df_prophet)

    # GÃ©nÃ©ration des dates futures jusqu'Ã  2024
    future_dates = model.make_future_dataframe(periods=12, freq='Y')
    forecast = model.predict(future_dates)

    return df_prophet, forecast

# Initialisation de l'application Dash
app = dash.Dash(__name__)

# Layout de l'application Dash
app.layout = html.Div([
    html.Label("SÃ©lectionnez une Route :"),
    dcc.Dropdown(
        id='route-dropdown',
        options=[{'label': f"Route {route}", 'value': route} for route in df_melted["Route"].unique()],
        value=df_melted["Route"].unique()[0],  # Route sÃ©lectionnÃ©e par dÃ©faut
        style={'width': '50%'}
    ),

    dcc.Graph(id='evolution-graph')
])

# Callback pour mettre Ã  jour le graphique en fonction de la route sÃ©lectionnÃ©e
@app.callback(
    Output('evolution-graph', 'figure'),
    [Input('route-dropdown', 'value')]
)
def update_graph(selected_route):
    df_prophet, forecast = predict_route(selected_route)

    fig = go.Figure()

    # Tracer les donnÃ©es observÃ©es
    fig.add_trace(go.Scatter(
        x=df_prophet["ds"],
        y=df_prophet["y"],
        mode='lines+markers',
        name=f"ObservÃ© - Route {selected_route}",
        hovertemplate='%{y} vÃ©hicules<br>AnnÃ©e: %{x|%Y}<extra></extra>',
    ))

    # Tracer les prÃ©dictions
    fig.add_trace(go.Scatter(
        x=forecast["ds"],
        y=forecast["yhat"],
        mode='lines',
        name=f"PrÃ©diction - Route {selected_route}",
        line=dict(dash='dash', color='red'),
        hovertemplate='%{y:.0f} vÃ©hicules<br>AnnÃ©e: %{x|%Y}<extra></extra>',
    ))

    # Ajouter l'incertitude (intervalle de confiance)
    fig.add_trace(go.Scatter(
        x=forecast["ds"].tolist() + forecast["ds"][::-1].tolist(),
        y=forecast["yhat_upper"].tolist() + forecast["yhat_lower"][::-1].tolist(),
        fill='toself',
        fillcolor='rgba(255, 0, 0, 0.2)',
        line=dict(color='rgba(255,255,255,0)'),
        hoverinfo="skip",
        showlegend=True,
        name="Incertitude",
    ))

    # Mise en page du graphique
    fig.update_layout(
        title=f'Ã‰volution et PrÃ©diction du trafic pour la Route {selected_route}',
        yaxis_title='Nombre de vÃ©hicules',
        hovermode='x unified',
        template='plotly_dark',
    )

    return fig

# Lancer l'application
if __name__ == '__main__':
    app.run_server(debug=True)
```
